```python
df['Date of Admission']=pd.to_datetime(df['Date of Admission'])

df['Discharge Date']=pd.to_datetime(df['Discharge Date'])
```

```python
#feature engineering

#Duration of stay in hospital
df['Duration of Stay']=(df['Discharge Date']-df['Date of Admission']).dt.days
```

```python
# Calculating the average duration of stay
average_duration_of_stay = df['Duration of Stay'].mean()
print("Average Duration of Hospital Stay:", average_duration_of_stay, "days")
```

#### Average Duration of Hospital Stay:

The average Duration of stay for patients is approximately **15.51 days**.

```python
#Duration of stay across different medical conditions

plt.figure(figsize=(10, 6))

sns.boxplot(data=df,x='Medical Condition', y='Duration of Stay', palette="Set3")

plt.xticks(rotation=90)

plt.title('Duration of Stay Across Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Duration of Stay (days)')
plt.show()
```

```python
#Average Duration of stay for each admission type

avg_duration_of_stay = df.groupby('Admission Type')['Duration of Stay'].mean().reset_index()

sns.barplot(x='Admission Type', y='Duration of Stay', data=avg_duration_of_stay, palette='husl')

plt.title('Average Duration of Stay by Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Average Duration of Stay (days)')
plt.show()
```

```python
# Distribution of blood types among different age groups

# Define age groups
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Plot the distribution of blood types among different age groups

plt.figure(figsize=(10, 6))

sns.countplot(data=df, x="Blood Type", hue="Age Group", palette="Set2")

plt.title('Blood Type Distribution Among Different Age Groups')
plt.xlabel('Blood Type')
plt.ylabel('Count')
plt.legend(title='Age Group', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```

```python
#Group by Medical Condition and calculate mean Billing Amount and Duration of Stay
condition_group = df.groupby('Medical Condition')[['Billing Amount', 'Duration of Stay']].mean().reset_index()

# Visualize the impact of Medical Condition on Billing Amount and Duration of Stay
plt.figure(figsize=(14, 6))
sns.barplot(x='Billing Amount', y='Medical Condition', data=condition_group, palette='viridis')
plt.title('Average Billing Amount by Medical Condition')
plt.xlabel('Average Billing Amount')
plt.ylabel('Medical Condition')
plt.show()

plt.figure(figsize=(14, 6))
sns.barplot(x='Duration of Stay', y='Medical Condition', data=condition_group, palette='viridis')
plt.title('Average Duration of Stay by Medical Condition')
plt.xlabel('Average Duration of Stay (Days)')
plt.ylabel('Medical Condition')
plt.show()
```

```python
bins = [0, 20, 30, 40, 50, 60, 70, 100]
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+']

# Assign age groups to a new column
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Iterate over each age group and plot
for group in df['Age Group'].unique():
    group_data = df[df['Age Group'] == group]
    
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=group_data['Age'], y=group_data['Billing Amount'])
    plt.title(f'Age vs Billing Amount for Age Group {group}')
    plt.xlabel('Age')
    plt.ylabel('Billing Amount')
    plt.show()
    
    # Calculate and print correlation for each group
    correlation = group_data['Age'].corr(group_data['Billing Amount'])
    print(f"Correlation between Age and Billing Amount for Age Group {group}: {correlation:.2f}")
```

```python
 #Perform the correlation analysis
correlation_matrix = df[['Age', 'Billing Amount', 'Duration of Stay']].corr()

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Display the correlation matrix
print(correlation_matrix)
```
